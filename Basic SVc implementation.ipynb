```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib as plt 
%matplotlib inline
import matplotlib.pyplot as plt
#lets create Synthetic Datasets
from sklearn.datasets import make_classification
```

```python
X,Y = make_classification(n_samples=1000,n_features=2,n_classes=2
                          ,n_clusters_per_class=1,n_redundant =0 )
```

```python
X
```

```python
Y
```

```python
#as this will give us first column
pd.DataFrame(X[0]), 
pd.DataFrame(X[1])
```

```python

sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=Y)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel= 'linear')
```

```python

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state = 10)
```

```python
svc.fit(X_train,Y_train)
```

```python
#predition
Y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print (classification_report(Y_test,Y_pred))
print(confusion_matrix(Y_test,Y_pred))
```

```python
#as our points are clearly seperable we get 100% accuracy and also get 100%accuracy if we use logistic regression
```

#now we try same data set with rbf kernel

```python
rbf = SVC(kernel= 'rbf')
```

```python

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state = 10)
```

```python
rbf.fit(X_train,Y_train)
```

```python
#predition
Y_pred = rbf.predict(X_test)
```

```python
print (classification_report(Y_test,Y_pred))
print(confusion_matrix(Y_test,Y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

#defining parameter range
param_grid = {'C':[0.1,1,10,100,1000],
             'gamma': [1,0.1,0.01,0.001,0.0001],
             'kernel':['rbf']}
```

```python
grid = GridSearchCV(estimator=SVC(), param_grid=param_grid, refit=True, cv=5, verbose=3)
```

```python
grid.fit(X_train,Y_train)
```

```python
grid.best_params_
```

```python
#predition
Y_pred = grid.predict(X_test)
print (classification_report(Y_test,Y_pred))
print(confusion_matrix(Y_test,Y_pred))
```
